```python
#from ensemble import RegressionEnsemble, MixedRegressionEnsemble, BootstrapEnsemble, ShuffleEnsemble
import ensemble
import importlib
importlib.reload(ensemble)

from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import RadiusNeighborsRegressor, KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.svm import SVR

from sklearn.ensemble import RandomForestRegressor

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
plt.rcParams['figure.figsize'] = [20, 5]

from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split

from sklearn.datasets import make_regression

np.random.seed(420)

num_experiments = 100

import warnings
warnings.filterwarnings('ignore')
```

# Linear Case

```python
X, y = make_regression(n_samples=200, n_features=1, noise=25,random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
plt.scatter(X,y)
```

```python
results_list = []

for i in range(num_experiments):
    np.random.seed(42+i)


    mixed_ensemble_2 = ensemble.MixedRegressionEnsemble(models = [#DecisionTreeRegressor(),
                                                                LinearRegression(),
                                                                #KNeighborsRegressor(),
                                                                #MLPRegressor(max_iter=1000),
                                                               SVR()
                                                                 ], 
                                                       seed = 42+i*800)
    mixed_regression = mixed_ensemble_2#DecisionTreeRegressor(max_depth=5)
    mixed_regression.fit(X, y)
    y_mixed,std_mixed = mixed_regression.predict(X_test,std=True)
    results = mixed_regression.self_evaluate(X_test,y_test)
    results_list.append(results)

    #mixed_ensemble_2.scatterplot(X_test,X=X,y=y)
df_tree_results = pd.DataFrame.from_records(results_list)

mixed_regression.mutli_dimenstional_scatterplot(X_test,y_test,figsize=(20,5))
```
